# Enunciado

Muestra cómo cargas el archivo train.csv de la competencia y muestras alguna exploración de datos (por ejemplo, cuantos estudiantes hay de cada clase, por localidades, etc.).
 
Esta exploración no tiene que ser exhaustiva, tan sólo que muestres que eres capaz de cargar los datos e inspeccionarlos.

# Solucion

## Carga de datos

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('./train.csv')
```

```python
dataset.tail(100)
```

## Exploración

### Numero de estudiantes por departamento

En el siguiente gráfico se muestra la cantidad de estudiantes por departamento en el dataset

```python
deparment = dataset.groupby('ESTU_PRGM_DEPARTAMENTO').size().sort_values(ascending=False)
```

```python
deparment.plot(kind='bar', figsize=(10, 6), color='skyblue')

plt.title('Cantidad de estudiantes por departamento')
plt.xlabel('Departamento')
plt.ylabel('Cantidad de estudiantes')
plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```

### Numero de estudiantes por programa

En el siguiente gráfico se muestran los 30 programas donde más estudiantes hay en el dataset

```python
program = dataset.groupby('ESTU_PRGM_ACADEMICO').size().sort_values(ascending=False).head(30)
```

```python
program.plot(kind='bar', figsize=(10, 6), color='skyblue')

plt.title('Cantidad de estudiantes por programa')
plt.xlabel('Programa')
plt.ylabel('Cantidad de estudiantes')
plt.xticks(rotation=90)
plt.tight_layout()
```

### Cantidad de estudiantes por ingeniería

En el siguiente gráfico aparecen los 30 programas de ingeniería con más estudiantes en el dataset

```python
engineeries = dataset[dataset['ESTU_PRGM_ACADEMICO'].str.startswith('ING')].groupby('ESTU_PRGM_ACADEMICO').size().sort_values(ascending=False).head(30)
```

```python
engineeries.plot(kind='bar', figsize=(10, 6), color='skyblue')

plt.title('Cantidad de estudiantes por programa')
plt.xlabel('Programa')
plt.ylabel('Cantidad de estudiantes')
plt.xticks(rotation=90)
plt.yticks(range(0, 30000, 5000))
plt.tight_layout()
```
